```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
blank_img = np.zeros((600,600))
```

```python
plt.imshow(blank_img)
```

```python
nphoto = cv2.putText(blank_img,
            text = "MANOJ",
            org=(50,300),
            fontFace = cv2.FONT_HERSHEY_SIMPLEX,
            fontScale = 5,
            color = (255,255,255),
            thickness = 20
           )
```

```python
plt.imshow(nphoto)
```

```python
plt.imshow(blank_img, cmap='gray')
```

```python
kernel = np.ones((5,5),dtype = np.uint8)
```

```python
kernel
```

```python
erode_img = cv2.erode(nphoto,kernel,iterations = 2)
```

```python
plt.imshow(erode_img)
```

```python
dilute_img = cv2.dilate(nphoto,kernel,iterations = 5)
```

```python
plt.imshow(dilute_img)
```

```python
gradient = cv2.morphologyEx(nphoto, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.imshow(gradient)
```

```python
opening = cv2.morphologyEx(nphoto, cv2.MORPH_OPEN, kernel)
```

```python
plt.imshow(opening)
```
